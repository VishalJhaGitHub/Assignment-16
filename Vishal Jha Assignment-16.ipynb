```python
#1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use MongoDB over SQL databases?

#Ans

#MongoDB is a popular document-oriented NoSQL database that allows users to store and retrieve data in a flexible and scalable manner. It uses a JSON-like document format to store data, which makes it easy to work with for developers and provides excellent performance for certain types of applications.

#Non-relational databases, also known as NoSQL databases, are databases that do not use a traditional relational data model with tables, columns, and rows. Instead, they store data in various formats such as documents, key-value pairs, or graph structures. Non-relational databases are often preferred over traditional relational databases in scenarios where data is unstructured, or when the data schema is subject to frequent changes. Non-relational databases can also provide excellent scalability and performance for certain types of applications.

#MongoDB is often preferred over SQL databases in scenarios where data is unstructured or subject to frequent changes. Because of its flexible document format, MongoDB is an excellent choice for storing data such as social media posts, product descriptions, or other types of unstructured data. MongoDB is also preferred for applications that require high scalability, as it can easily scale horizontally by adding additional nodes to a cluster. Additionally, MongoDB provides built-in support for sharding and replica sets, which can help ensure high availability and data durability. Finally, MongoDB is a good fit for applications that require real-time data analysis or stream processing, as it provides native support for aggregation pipelines and change streams.
```

```python
#2. State and Explain the features of MongoDB.

#Ans

#1 - Document-oriented data model: MongoDB uses a document-oriented data model to store data, which means that data is organized into documents rather than tables with rows and columns. This allows for more flexibility in data structures and makes it easier to represent complex data relationships.

#2 - Dynamic schema: MongoDB has a dynamic schema, which means that documents in a collection can have different fields and structures. This allows for easier data modeling and faster iterations in development.

#3 - High performance: MongoDB is designed for high performance and can handle large volumes of data and high throughput. It also has built-in caching mechanisms that improve read performance.

#4 - Horizontal scalability: MongoDB is horizontally scalable, which means that it can scale out across multiple servers to handle larger workloads. This is achieved through sharding, which distributes data across multiple nodes.

#5 - Querying and indexing: MongoDB supports powerful query language and indexing options, including geospatial indexing and text search. This makes it easy to search for and retrieve specific data in a collection.

#6 - Replication and high availability: MongoDB supports automatic replication and failover, which means that it can ensure high availability and data redundancy in case of hardware or software failures.

#7 - Easy integration: MongoDB has extensive drivers and APIs that make it easy to integrate with a variety of programming languages and frameworks.
```

```python
#4. Using the database and the collection created in question number 3, write a code to insert one record, and insert many records. Use the find() and find_one() methods to print the inserted record.

#Ans

# import the pymongo library
import pymongo

# connect to the MongoDB server running on localhost
client = pymongo.MongoClient("mongodb://localhost:27017/")
# create a new database named "mydatabase"
db = client["mydatabase"]
# create a new collection named "customers" in the "mydatabase" database
col = db["customers"]
# insert one record into the "customers" collection
doc = {"name": "Vishal Jha", "email": "vishaljha@example.com"}
x = col.insert_one(doc)
print("Inserted ID:", x.inserted_id)
# insert many records into the "customers" collection
docs = [
    {"name": "ABC", "email": "abc@example.com"},
    {"name": "DEF", "email": "def@example.com"},
    {"name": "GHI", "email": "ghi@example.com"}
]
x = col.insert_many(docs)
print("Inserted IDs:", x.inserted_ids)
# find the inserted record with name "Vishal Kumar"
result = col.find_one({"name": "Vishal Kumar"})
print("One record:", result)
# find all inserted records
results = col.find()
print("All records:")
for r in results:
    print(r)
```

```python
#5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to demonstrate this.

#Ans

#The find() method in MongoDB is used to query a collection and retrieve the documents that match the specified criteria. The method returns a cursor object, which allows you to iterate over the results.

# import the pymongo library
import pymongo
# connect to the MongoDB server running on localhost
client = pymongo.MongoClient("mongodb://localhost:27017/")
# access the "mydatabase" database and "customers" collection
db = client["mydatabase"]
col = db["customers"]
# find all documents in the collection
results = col.find()
# iterate over the cursor object and print the results
for r in results:
    print(r)
```

```python
#6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

#Ans

#The sort() method in MongoDB is used to sort the documents in a collection based on one or more fields. The method takes one or more key-value pairs that specify the fields to sort by and the sort order (ascending or descending).

# import the pymongo library
import pymongo
# connect to the MongoDB server running on localhost
client = pymongo.MongoClient("mongodb://localhost:27017/")
# access the "mydatabase" database and "users" collection
db = client["mydatabase"]
col = db["users"]
# find all documents in the collection and sort by the "age" field in ascending order
results = col.find().sort("age", 1)
# print the sorted results
for r in results:
    print(r)
```

```python
#7. Explain why delete_one(), delete_many(), and drop() is used.

#Ans

#delete_one(), delete_many(), and drop() are used in MongoDB to remove data from a collection.

#1 - delete_one() method removes a single document that matches the specified filter criteria. If multiple documents match the criteria, only the first one encountered will be deleted.

#2 - delete_many() method removes all the documents that match the specified filter criteria.

#3 - The drop() method, on the other hand, removes an entire collection from the database. It does not take any arguments and deletes the entire collection.
```
